```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
K = np.loadtxt("therm.txt")
plt.plot(K.T[0],K.T[1],'k--',label = '90 K')
plt.plot(K.T[0],K.T[2],'g--',label = '80 K')
plt.plot(K.T[0],K.T[3],'r--',label = '70 K')
plt.legend()
plt.title("Thermal Conductivity (k) in W/mK ")
plt.xlabel("Correlation time (t) in ps")
plt.ylabel("k in W/mK")
plt.xlim(0,2)
plt.savefig("Thermal.png", dpi = 300)
plt.show()
```
